## **Exemplo** : Regressão

```python
# Importando bibliotecas.

import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt


from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Dataset Vizualization
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)
diabetes_X

### Observação: cada uma das colunas já está normalizada ###
```

```python
diabetes_y
```

```python
# Selecionando apenas uma variável para a Regressão Linear
diabetes_X = diabetes_X[:, np.newaxis, 2]

# Separando os dados para treinamento e teste
diabetes_X_train = diabetes_X[:-20]
diabetes_X_test = diabetes_X[-20:]

diabetes_y_train = diabetes_y[:-20]
diabetes_y_test = diabetes_y[-20:]

# ___?___ o modelo de regressor linear
regr = linear_model.LinearRegression()

# ___?___ o modelo de regressor linear
regr.fit(diabetes_X_train, diabetes_y_train)

# Fazendo as predições com o modelo de Regressão Linear
diabetes_y_pred = regr.predict(diabetes_X_test)

# Calculando o erro do modelo
erro = mean_squared_error(diabetes_y_test, diabetes_y_pred)/100
print("erro em termos percentuais:","{0:.2f}".format(erro))
```

```python
print("Valor correto: ", diabetes_y_test[0])
print("valor predição: ", diabetes_y_pred[0])
```

```python
# Plotando o gráfico do nosso resultado

plt.figure(figsize=(10, 7))
plt.scatter(diabetes_X_test, diabetes_y_test, color="black")
plt.plot(diabetes_X_test, diabetes_y_pred, color="purple", linewidth=2)
plt.xticks(())
plt.yticks(())
plt.show()

```

```python
plt.figure(figsize=(10, 7))
sns.regplot(data=diabetes_X_test, x=diabetes_y_pred, y=diabetes_y_test,line_kws={"color": "red"})
plt.show()
```
